```python
import yfinance as yf   # yfinance makes it easy to get stock data.
stock = yf.Ticker("AAPL")  # Using Apple's stock symbol
data = stock.history(period="1y")  # Fetching data for 1 year
```

```python
data.dropna(inplace=True)  # This removes any rows with missing values

data
```

```python

data.index = pd.to_datetime(data.index)  #ensure date format is correct.

data
```

```python

data['Daily Return'] = data['Close'].pct_change()  # to calculate daily returns

data['Daily Return']
```

```python
import matplotlib.pyplot as plt

data['Close'].plot(figsize=(12, 6), title="Stock Price Movement")        # To Visualize stock price movements.
plt.show()
```

```python

data['Volume'].plot(figsize=(12, 6), title="Volume Traded Over Time")  # Analyze volume of stocks traded.
plt.show()
```

```python
import seaborn as sns

stock_list = ["AAPL", "MSFT", "JPM", "PEP", "GOOGL"]     # Correlation between stocks.
                                                         # Correlation values range between -1 and 1.
close_prices = pd.DataFrame()                            #  A value closer to 1 implies that two stocks move in tandem, while a value closer to -1 indicates they move in opposite directions.

for stock in stock_list:
   close_prices[stock] = yf.Ticker(stock).history(period="1y")['Close']

correlation = close_prices.pct_change().corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.show()
```

```python
data['SMA50'] = data['Close'].rolling(window=50).mean()      # Simple moving average
data['SMA200'] = data['Close'].rolling(window=200).mean()    # Here, SMA50 is the 50-day moving average, and SMA200 is the 200-day moving average.
                                                             # When SMA50 crosses above SMA200, it's often seen as a bullish sign, and vice versa.
data[['Close', 'SMA50', 'SMA200']].plot(figsize=(12,6))
plt.title('Simple Moving Average')
plt.show()
```

```python
data['EMA50'] = data['Close'].ewm(span=50, adjust=False).mean()      # Exponential Moving Average
data['EMA200'] = data['Close'].ewm(span=200, adjust=False).mean()    # EMA gives more weight to recent prices, and therefore reacts more quickly to price changes than SMA.

data[['Close', 'EMA50', 'EMA200']].plot(figsize=(12,6))
plt.title('Exponential Moving Average')
plt.show()
```

```python
data['SMA20'] = data['Close'].rolling(window=20).mean()                   # Bolinger Band
data['Upper'] = data['SMA20'] + 2*data['Close'].rolling(window=20).std()
data['Lower'] = data['SMA20'] - 2*data['Close'].rolling(window=20).std()

data[['Close', 'SMA20', 'Upper', 'Lower']].plot(figsize=(12,6))
plt.title('Bollinger Bands')
plt.show()
```

```python
from scipy.optimize import minimize

# Assuming stock_list contains the symbols of the stocks in the portfolio
returns = close_prices.pct_change()
mean_returns = returns.mean()
cov_matrix = returns.cov()
num_portfolios = 10000
risk_free_rate = 0.0178 # example value

# Portfolio optimization function
def portfolio_annualised_performance(weights, mean_returns, cov_matrix):
   returns = np.sum(mean_returns*weights ) *252
   std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) * np.sqrt(252)
   return std, returns

# Minimize negative Sharpe Ratio to get optimal portfolio
def neg_sharpe_ratio(weights, mean_returns, cov_matrix, risk_free_rate):
   p_var, p_ret = portfolio_annualised_performance(weights, mean_returns, cov_matrix)
   return -(p_ret - risk_free_rate) / p_var

# Constraints for optimization
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

# Running the optimization
initial = [1./len(stock_list) for stock in stock_list]
bounds = tuple((0, 1) for asset in range(len(stock_list)))
result = minimize(neg_sharpe_ratio, initial, args=(mean_returns, cov_matrix, risk_free_rate), bounds=bounds, constraints=constraints)

```
